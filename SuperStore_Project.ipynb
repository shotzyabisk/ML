```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('train.csv')
df
```

```python
{i : len(df[i].unique()) for i in df.columns}
```

```python
df.info()
```

```python
df1 = df.copy()
```

```python
df1 = df1.drop(['Row ID','Order ID','Country','Product Name','Customer Name','Customer ID'],axis = 1)
df1
```

```python
df1['Order Date'] = pd.to_datetime(df1['Order Date'],format='%m/%d/%y', errors='coerce')
df1['Order Date']
```

```python
df1['Order_Date'+'_year'] = df1['Order Date'].apply(lambda x: x.year)
df1['Order_Date'+'_month'] = df1['Order Date'].apply(lambda x: x.month)
df1['Order_Date'+'_day'] = df1['Order Date'].apply(lambda x: x.day)
df1 = df1.drop(['Order Date'],axis =1)
df1
```

```python
df1['Ship Date'] = pd.to_datetime(df1['Ship Date'],format='%m/%d/%y', errors='coerce')
df1['Ship Date'+'_year'] = df1['Ship Date'].apply(lambda x: x.year)
df1['Ship Date'+'_month'] = df1['Ship Date'].apply(lambda x: x.month)
df1['Ship Date'+'_day'] = df1['Ship Date'].apply(lambda x: x.day)
df1 = df1.drop(['Ship Date'],axis =1)
df1
```

```python
def onehot_encode(df1,column):
  df1 = df1.copy()
  dummies = pd.get_dummies(df1[column],prefix= column)
  df1 = pd.concat([df1,dummies],axis  = 1)
  df1 = df1.drop(column,axis = 1)
  return df1
```

```python
for i in ['Ship Mode','Segment','City','State','Postal Code','Region','Product ID','Category','Sub-Category']:
  df1 =onehot_encode(df1,column = i)
df1
```

```python

df1 = df1.replace(np.nan, 0)
```

```python
y = df1['Sales']
x = df1.drop('Sales',axis = 1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.7,shuffle  = True, random_state = 1)
```

```python
x_train.describe()
#since the mean and standard deviation has a lot of variations we scale the data
```

```python
#standardscaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train  = pd.DataFrame(scaler.transform(x_train),columns = x.columns)
x_test  = pd.DataFrame(scaler.transform(x_test),columns = x.columns)
```

```python
x_train.describe()
```

```python
#model.add(keras.Input(shape=(data.shape[1], )))
inputs = tf.keras.Input(shape  = (x_train.shape[1],))
x1 = tf.keras.layers.Dense(384, activation = 'relu')(inputs)
x1 = tf.keras.layers.Dense(384, activation = 'relu')(x1)
outputs  = tf.keras.layers.Dense(1,activation = 'linear')(x1)
model = tf.keras.Model(inputs = inputs , outputs = outputs)
print(model.summary())
```

```python
model.compile(optimizer = 'adam',loss = 'mse')
history = model.fit(
    x_train,
    y_train ,
    validation_split = 0.2, batch_size =32,epochs = 100,
    callbacks =  [tf.keras.callbacks.EarlyStopping(monitor= 'val_loss',patience = 5,restore_best_weights = True),
                  tf.keras.callbacks.ReduceLROnPlateau()
                  ]
    )
```

```python
test_loss = model.evaluate(x_test , y_test,verbose = 0)
print(test_loss)
```

```python
y_pred = np.squeeze(model.predict(x_test))
```

```python
r2_score(y_test,y_pred)
```
